```python
# Dependencies
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as st
import numpy as np
import altair as alt
#from matplotlib import * as plt

who_data = pd.read_csv('Resources/WHO_data.csv')
#removing nulls
who_data.dropna(inplace=True)
#we can use fillna and replace the values with mean or modes or anything you want
```

```python
plt.scatter(who_data.iloc[:,1],who_data.iloc[:,8])
plt.xlabel('Income_Per_Capita')
plt.ylabel('Average Alcohol Consumed Per Person Per Year (L)')
plt.show()
income =who_data.iloc[:,1]
alcohol = who_data.iloc[:,8]
correlation = st.pearsonr(income,alcohol)
print(f"r = {round(correlation[0],2)}")
```
