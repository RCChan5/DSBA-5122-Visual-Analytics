#### Using the following data, plot the monthly averages for temperature in New Jersey in both degrees Fahrenheit and degrees Celsius.

#### Average temperature per month in Fahrenheit: 39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44.

```python
#import dependencies
import random
import numpy as np
import matplotlib.pyplot as plt


avgTempF = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
print(avgTempF)
```

#### Assign to the x-axis a range of numerical values representing each month of the year.

```python
x=[i for i in range(1,13)]
print(x)
```

#### Plot the degrees Fahrenheit points.

```python
plt.plot(x,avgTempF)
plt.xlabel("Month")
plt.ylabel("Fahrenheit")
plt.title("Fahrenheit v Month")
```

#### Use a list comprehension to convert the temperature to degrees Celsius.

```python
avgTempC = [(F-32)*.56 for F in avgTempF]
print(avgTempC)
print(type(avgTempC[1]))
```

#### Plot the degrees Celsius points.

```python
plt.plot(x,avgTempC)
plt.xlabel("Month")
plt.ylabel("Celsius")
plt.title("Celsius v Month")
```

#### Create a third plot with both the degrees Fahrenheit and degrees Celsius points.

```python
plt.plot(x,avgTempF, label = "Average Temp in Fahrenheit")
plt.plot(x,avgTempC, label = "Average Temp in Celsius")
plt.xlabel("Month")
plt.ylabel("degrees")
plt.title("double line graph of avg temp F v C")
plt.legend()
```
